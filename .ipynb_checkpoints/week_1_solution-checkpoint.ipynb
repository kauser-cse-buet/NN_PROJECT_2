```python
import urllib2
import tarfile

# download data 
# https://github.com/mnielsen/neural-networks-and-deep-learning/blob/master/data/mnist.pkl.gz

url = "https://github.com/mnielsen/neural-networks-and-deep-learning/blob/master/data/mnist.pkl.gz"
mnist_gz_file = urllib2.urlopen(url)

mnist_local_file = "mnist.pkl.gz"
with open(mnist_local_file,'wb') as output:
    output.write(mnist_gz_file.read())
print("# File downloaded: ")
      
tar = tarfile.open(mnist_local_file, "r:gz")
tar.extractall()
tar.close()
```
