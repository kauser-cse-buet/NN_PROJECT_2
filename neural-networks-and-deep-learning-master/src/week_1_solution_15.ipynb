```python
import mnist_loader
import numpy as np 
import pandas as pd 

training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
print("Training data, validation data and test data loaded.")
```

```python
print("15. Reapet 12 for a deep network with four hidden layers, with 30 neuron in each.")
```

```python
import network2
net2_all_train = network2.Network([784, 30, 30, 30, 30, 10], cost=network2.CrossEntropyCost)
net2_all_train.large_weight_initializer()

epochs = 30

evaluation_cost_all_train, evaluation_accuracy_all_train, training_cost_all_train, training_accuracy_all_train = net2_all_train.SGD(
    training_data, 
    epochs, 
    10, 
    0.5, 
    evaluation_data=test_data,
    lmbda=5.0,
    monitor_evaluation_cost=True,
    monitor_evaluation_accuracy=True,
    monitor_training_cost=True,
    monitor_training_accuracy=True
)
```

```python
import plot_helper

plot_helper.plot_training_cost(num_epochs=epochs, training_cost=training_cost_all_train, training_cost_xmin=0)
plot_helper.plot_training_accuracy(num_epochs=epochs, training_accuracy=training_accuracy_all_train, training_accuracy_xmin=0, training_set_size=10000)

plot_helper.plot_test_cost(num_epochs=epochs, test_cost=evaluation_cost_all_train, test_cost_xmin=0)
plot_helper.plot_test_accuracy(num_epochs=epochs, test_accuracy=evaluation_accuracy_all_train, test_accuracy_xmin=0)
```
